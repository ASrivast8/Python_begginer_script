```python
################################################################################################################################################
#              List                ##              List                ##              List                ##              List                #
################################################################################################################################################


# 1. Reverse a List
def reverse_list(nums):
    return nums[::-1]

nums = [3, 1, 4, 1, 5, 9]
print(reverse_list(nums))  # Output: [9, 5, 1, 4, 1, 3]

# 2. Find the Largest and Smallest Elements in a List
def find_largest_and_smallest(nums):
    return max(nums), min(nums)

nums = [3, 1, 4, 1, 5, 9]
print(find_largest_and_smallest(nums))  # Output: (9, 1)

# 3. Remove Duplicates from a List
def remove_duplicates(nums):
    return list(set(nums))

nums = [1, 2, 3, 2, 4, 4, 5]
print(remove_duplicates(nums))  # Output: [1, 2, 3, 4, 5]

# 4. Find the Intersection of Two Lists
def intersection(list1, list2):
    return list(set(list1) & set(list2))

list1 = [1, 2, 3, 4]
list2 = [3, 4, 5, 6]
print(intersection(list1, list2))  # Output: [3, 4]

# 5. Sum of All Numbers in a List
def sum_list(nums):
    return sum(nums)

nums = [3, 1, 4, 1, 5, 9]
print(sum_list(nums))  # Output: 23

# 6. Find the Index of the First Occurrence of an Element
def find_index(nums, target):
    return nums.index(target) if target in nums else -1

nums = [3, 1, 4, 1, 5, 9]
print(find_index(nums, 4))  # Output: 2

# 7. Flatten a Nested List
def flatten_list(nested_list):
    return [item for sublist in nested_list for item in sublist]

nested_list = [[1, 2], [3, 4], [5, 6]]
print(flatten_list(nested_list))  # Output: [1, 2, 3, 4, 5, 6]

# 8. Find the Most Frequent Element in a List
from collections import Counter
def most_frequent(nums):
    counter = Counter(nums)
    return counter.most_common(1)[0][0]

nums = [3, 1, 4, 1, 5, 9, 1]
print(most_frequent(nums))  # Output: 1

# 9. Find All Pairs in a List that Sum to a Target Value

def find_pairs(nums, target):
    pairs = []
    
    for i in range(len(nums)):
        for j in range(i+1, len(nums)):
            if nums[i] + nums[j] == target:
                pairs.append((nums[i], nums[j]))
                
    return pairs



def find_pairs(nums, target):  #(nums = [3, 1, 4, 1, 5, 9],8)
    pairs = []
    seen = set()
    for num in nums:
        complement = target - num
        if complement in seen:
            pairs.append((complement, num))
        seen.add(num)
    return pairs

nums = [3, 1, 4, 1, 5, 9]
print(find_pairs(nums, 6))  # Output: [(5, 1), (4, 2)]

# 10. Rotate a List by N Places
def rotate_list(nums, n):
    return nums[-n:] + nums[:-n]

nums = [1, 2, 3, 4]
print(rotate_list(nums, 2))  # Output: [3, 4, 1, 2]





################################################################################################################################################
#              Dictionary          ##              Dictionary          ##              Dictionary          ##              Dictionary          #
################################################################################################################################################

# 1. Merge Two Dictionaries
def merge_dicts(dict1, dict2):
    dict1.update(dict2)
    return dict1

dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}
print(merge_dicts(dict1, dict2))  # Output: {'a': 1, 'b': 2, 'c': 3, 'd': 4}

# 2. Find the Key with the Maximum Value
def max_key_value(d):
    return max(d, key=d.get)

d = {'a': 1, 'b': 3, 'c': 2}
print(max_key_value(d))  # Output: b

# 3. Count the Frequency of Each Character in a String
def count_char_frequency(s):
    return {char: s.count(char) for char in set(s)}

s = "hello"
print(count_char_frequency(s))  # Output: {'h': 1, 'e': 1, 'l': 2, 'o': 1}

# 4. Invert a Dictionary
def invert_dict(d):
    return {v: k for k, v in d.items()}

d = {'a': 1, 'b': 2, 'c': 3}
print(invert_dict(d))  # Output: {1: 'a', 2: 'b', 3: 'c'}

# 5. Check if Two Dictionaries Are Equal
def are_dicts_equal(dict1, dict2):
    return dict1 == dict2

dict1 = {'a': 1, 'b': 2}
dict2 = {'a': 1, 'b': 2}
print(are_dicts_equal(dict1, dict2))  # Output: True

# 6. Get the Keys with a Specific Value
def get_keys_with_value(d, target_value):
    return [k for k, v in d.items() if v == target_value]

d = {'a': 1, 'b': 2, 'c': 1}
print(get_keys_with_value(d, 1))  # Output: ['a', 'c']


data = {'a': 10, 'b': 20, 'c': 10, 'd': 30}
target = 10

# filter returns an iterator; we convert it to a list
keys = list(filter(lambda k: data[k] == target, data))

print(keys)
# Output: ['a', 'c']

# 7. Remove a Key-Value Pair from a Dictionary
def remove_key(d, key):
    if key in d:
        del d[key]
    return d

d = {'a': 1, 'b': 2, 'c': 3}
print(remove_key(d, 'b'))  # Output: {'a': 1, 'c': 3}

# 8. Find the Common Keys Between Two Dictionaries
def common_keys(dict1, dict2):
    return list(dict1.keys() & dict2.keys())

dict1 = {'a': 1, 'b': 2}
dict2 = {'b': 3, 'c': 4}
print(common_keys(dict1, dict2))  # Output: ['b']

# 9. Merge a List of Dictionaries
def merge_dict_list(dict_list):
    result = {}
    for d in dict_list:
        result.update(d)
    return result

dict_list = [{'a': 1}, {'b': 2}, {'c': 3}]
print(merge_dict_list(dict_list))  # Output: {'a': 1, 'b': 2, 'c': 3}

# 10. Sort a Dictionary by Values
def sort_dict_by_value(d):
    return dict(sorted(d.items(), key=lambda item: item[1]))

d = {'a': 1, 'b': 3, 'c': 2}
print(sort_dict_by_value(d))  # Output: {'a': 1, 'c': 2, 'b': 3}

# 11.Combine dictionary keys using union and intersection.
from collections import Counter

data = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
counts = Counter(data)

print(counts)
# Output: Counter({'apple': 3, 'banana': 2, 'orange': 1})

# Bonus: Get the most frequent item easily
print(counts.most_common(1)) 
# Output: [('apple', 3)]




################################################################################################################################################
#              string          ##              string          ##              string          ##              string                          #
################################################################################################################################################

# 1. Reverse Words in a Sentence
def reverse_words(s):
    return " ".join(s.split()[::-1])

print(reverse_words("Hello World Python"))  # Output: "Python World Hello"


# 2. Check if a String is a Permutation of Another
from collections import Counter

def is_permutation(s1, s2):
    return Counter(s1) == Counter(s2)

print(is_permutation("listen", "silent"))  # Output: True

# Generate Permutations in Python

from itertools import permutations

s = "abc"

result = permutations(s)

print(list(result))

["".join(p) for p in permutations(s)]




# 3. Find the First Non-Repeating Character
def first_unique_char(s):
    char_count = Counter(s)
    for char in s:
        if char_count[char] == 1:
            return char
    return None

print(first_unique_char("swiss"))  # Output: 'w'


# 4. Longest Common Prefix
def longest_common_prefix(words):
    if not words: return ""
    prefix = words[0]
    for word in words[1:]:
        while not word.startswith(prefix):
            prefix = prefix[:-1]
            if not prefix:
                return ""
    return prefix

print(longest_common_prefix(["flower", "flow", "flight"]))  # Output: "fl"


# 5. Count Occurrences of Each Character
def char_count(s):
    return dict(Counter(s))

print(char_count("hello"))  # Output: {'h': 1, 'e': 1, 'l': 2, 'o': 1}


# 6. Check if Two Strings are One Edit Away
def one_edit_away(s1, s2):
    if abs(len(s1) - len(s2)) > 1:
        return False
    from itertools import zip_longest
    diffs = sum(c1 != c2 for c1, c2 in zip_longest(s1, s2, fillvalue=""))
    return diffs <= 1

print(one_edit_away("pale", "ple"))  # Output: True


# 7. Compress a String (Run-Length Encoding)
def run_length_encoding(s):
    if not s: return ""
    encoded, count = [], 1
    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            count += 1
        else:
            encoded.append(s[i - 1] + str(count))
            count = 1
    encoded.append(s[-1] + str(count))
    return "".join(encoded)

print(run_length_encoding("aaabbc"))  # Output: "a3b2c1"


# 8. Find the Longest Palindromic Substring
def longest_palindrome(s):
    def expand_around_center(left, right):
        while left >= 0 and right < len(s) and s[left] == s[right]:
            left -= 1
            right += 1
        return s[left+1:right]
    
    result = ""
    for i in range(len(s)):
        for temp in (expand_around_center(i, i), expand_around_center(i, i+1)):
            if len(temp) > len(result):
                result = temp
    return result

print(longest_palindrome("babad"))  # Output: "bab" or "aba"


def all_palindromes(s):
    results = []
    for i in range(len(s)):
        for l, r in [(i, i), (i, i + 1)]:
            while l >= 0 and r < len(s) and s[l] == s[r]:
                results.append(s[l:r+1])
                l -= 1
                r += 1
    return results


# 9. Find the Most Frequent Word in a String
def most_frequent_word(s):
    words = s.split()
    return max(Counter(words), key=Counter(words).get)

print( most_frequent_word("apple banana apple orange banana apple") )  # Output: "apple"


# 10. Check if a String Follows a Given Pattern
def word_pattern(pattern, s):
    words = s.split()
    if len(pattern) != len(words):
        return False
    mapping, used_words = {}, set()
    for p, w in zip(pattern, words):
        if p in mapping:
            if mapping[p] != w:
                return False
        elif w in used_words:
            return False
        else:
            mapping[p] = w
            used_words.add(w)
    return True

print(word_pattern("abba", "dog cat cat dog"))  # Output: True

```

```python
def longest_palindrome(s):
    start, end = 0, 0
    def expand_around_center(left, right):
        while left >= 0 and right < len(s) and s[left] == s[right]:
            left -= 1
            right += 1
        return left + 1, right - 1 

    for i in range(len(s)):
        l1, r1 = expand_around_center(i, i)    
        l2, r2 = expand_around_center(i, i + 1)
        if r1 - l1 > end - start:
            start, end = l1, r1
        if r2 - l2 > end - start:
            start, end = l2, r2

    return s[start:end + 1]


s = 'adadadada'
longest_palindrome(s)
```

```python

def longest_common_prefix(words):
    if not words: return ""
    prefix = words[0]
    for word in words[1:]:
        while not word.startswith(prefix):
            prefix = prefix[:-1]
            if not prefix:
                return ""
    return prefix

print(longest_common_prefix(["flower", "flow", "flight"]))  # Output: "fl"
```

```python
def all_palindromes(s):
    results = []
    for i in range(len(s)):
        for l, r in [(i, i), (i, i + 1)]:
            while l >= 0 and r < len(s) and s[l] == s[r]:
                results.append(s[l:r+1])
                l -= 1
                r += 1
    return results

all_palindromes("babad")    
```

```python
#  Find the Longest Palindromic Substring
def longest_palindrome(s):
    def expand_around_center(left, right):
        while left >= 0 and right < len(s) and s[left] == s[right]:
            left -= 1
            right += 1
        return s[left+1:right]
    
    result = ""
    for i in range(len(s)):
        for temp in (expand_around_center(i, i), expand_around_center(i, i+1)):
            if len(temp) > len(result):
                result = temp
    return result

print(longest_palindrome("babad"))  # Output: "bab" or "aba"
```

```python
# Merge

def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    sorted_arr = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            sorted_arr.append(left[i])
            i += 1
        else:
            sorted_arr.append(right[j])
            j += 1
    sorted_arr.extend(left[i:])
    sorted_arr.extend(right[j:])
    return sorted_arr

print(merge_sort([5, 3, 8, 1, 2]))
```
